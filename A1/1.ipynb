```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the dataset using Pandas
data1 = pd.read_csv("/content/drive/MyDrive/all_six_datasets/all_six_datasets/ETT-small/ETTh1.csv")
```

```python
data1.head()
```

```python
data1.describe()
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Create a plot of 'date' vs 'OT'
plt.figure(figsize=(12, 6))
plt.plot(data1['date'], data1['OT'])
plt.xlabel('Date')
plt.ylabel('OT')
plt.title('Date vs OT')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Set up the subplots
fig, axs = plt.subplots(2, 3, figsize=(18, 12))
attributes = ['HUFL', 'HULL', 'MUFL', 'MULL', 'LUFL', 'LULL', 'date']

for i, ax in enumerate(axs.flatten()):
    attr = attributes[i]
    if attr == 'date':
        continue  # Skip 'date' in the plots
    ax.plot(data1[attr], data1['OT'])
    ax.set_xlabel(attr)
    ax.set_ylabel('OT')
    ax.set_title(f'{attr} vs OT')

# Adjust the layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Create a plot of 'OT' only
plt.figure(figsize=(12, 6))
plt.plot(data1['OT'])
plt.xlabel('Data Point Index')
plt.ylabel('OT')
plt.title('OT Plot')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
